```python
# 각 채널의 정보를 저장하는 함수
def func_get_info(ch_list):
    ret_list = []

    for channel in ch_list:
        category = channel.select('.category')[0].text.strip()
        title = channel.select('h1>a')[0].text.strip()
        subscriber = channel.select('.subscriber_cnt')[0].text.strip()
        view = channel.select('.view_cnt')[0].text.strip()
        video = channel.select('.video_cnt')[0].text.strip()

    #     print(category, title, subscriber, view, video)
        data = [category, title, subscriber, view, video]
        ret_list.append(data)

    return ret_list
    
```

```python
page = 1
url = f'https://youtube-rank.com/board/bbs/board.php?bo_table=youtube={page}'
print(url)
```

```python
# 10개 페이지에서 채널 데이터 가져오기
import time

results = []

for page in range(1, 11):  # 1 ~ 10
    # url에서 페이지를 불러오기
    url = f'https://youtube-rank.com/board/bbs/board.php?bo_table=youtube&page={page}'
    driver.get(url)
    
    # 가져오는 동안 잠시 기다리기
    time.sleep(2)
    
    # html 파싱
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    
    # 채널 리스트 추출
    ch_list = soup.select('form>table>tbody>tr')
    
    # 각 정보 추출
    results += func_get_info(ch_list)
    
pd.DataFrame(results)
```

```python
df = pd.read_excel('youtube_rank_20230704_213832.xlsx')
```

```python
df.head()
```

```python
df['구독자수_int'] = df['구독자수'].str.replace('만','')
```

```python
df['구독자수_int'] = df['구독자수_int'].astype('int')
df.info()
```

```python
df_gb = df.groupby('카테고리').agg({'구독자수_int' : ['sum','count']})
df_gb.head()
```

```python
df_pt = df.pivot_table(index = '카테고리'
                      , values='구독자수_int'
                      , aggfunc=['sum','count'])
df_pt.head()
```

```python
#칼럼명 변경
df_gb.columns = ['구독자수_합','채널개수']
df_gb.head()
```

```python
#인데스 초기화
df_gb = df_gb.reset_index()
df_gb.head()
```

```python
# 구독자수_합 내림차순 정렬
df_gb = df_gb.sort_values(by='구독자수_합',ascending=False)
```

```python
# 시각화
import matplotlib.pyplot as plt

# 한글, - 기호 깨짐 방지
plt.rc('font', family='AppleGothic')
plt.rcParams['axes.unicode_minus'] = False
```

```python
#카테고리별 구독자수 합
plt.pie(df_gb['구독자수_합'],labels=df_gb['카테고리'],autopct='%.2f%%')
plt.show()
```

```python
#상위 5개 카테고리 + 5개를 제외한 나머지 카테고리의 합 테이블
df_top5 = df_gb[:5]
df_top5
```

```python
etc_sub_sum = df_gb[5:]['구독자수_합'].sum()
etc_ch_sum = df_gb[5:]['채널개수'].sum()

print(etc_sub_sum)
print(etc_ch_sum)
```

```python
#[기타] 카테고리로 데이터프레임을 만들고 top5 테이블과 합치기
etc_data = {'카테고리':'[기타]',
           '구독자수_합':[etc_sub_sum],
           '채널개수':[etc_ch_sum]}

df_etc = pd.DataFrame(etc_data)

```

```python
df_top5 = pd.concat([df_top5, df_etc])
df_top5
```

```python
plt.figure(figsize=(4,4))
plt.pie(df_top5['구독자수_합']
       ,labels=df_top5['카테고리']
        ,autopct = '%.2f%%' #소숫점 둘째자리 표시
        ,counterclock = False #시계방향 표시
        ,startangle = 90 #시작점을 90도 회전
       )

plt.show()
```

```python
#카테고리별 채널 개수 
plt.figure(figsize=(4,4))
plt.pie(df_top5['채널개수']
       ,labels=df_top5['카테고리']
        ,autopct = '%.1f%%' #소숫점 첫째자리 표시
        ,counterclock = False #시계방향 표시
        ,startangle = 90 #시작점을 90도 회전
       )

plt.show()
```
